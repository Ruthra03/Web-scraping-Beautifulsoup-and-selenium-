```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
from google.colab import files
```

```python
# Function to extract company name
def get_company_name(soup):

    try:
        Company_name = soup.find('div', class_='heading_6 company_name').text.strip()

    except AttributeError:
        Company_name = ""

    return Company_name

def get_role(soup):

    try:
        Role = soup.find("h1", attrs={"class":'heading_2_4 heading_title'}).text.strip()

    except AttributeError:
        Role = ""

    return Role

def get_location(soup):

    try:
        span_elements = soup.find_all('span')
        # Keep track of the count of <span> elements containing <a> tags
        count = 0
        for span in span_elements:
          if span.find('a'):
            count += 1
          if count == 3:
            # Extract the text content from the <a> tag inside the second <span>
            location = span.a.text.strip()
            break

    except AttributeError:
        location = ""
    return location

def get_last_date(soup):

    try:
        div_tags_with_single_class = [div for div in new_soup.find_all('div', class_='item_body', id=False) if len(div['class']) == 1]

        for div_tag in div_tags_with_single_class:
          last_date = div_tag.text.strip()

    except AttributeError:
        last_date = ""

    return last_date

def get_salary(soup):

    try:
        salary = new_soup.find('span', class_='desktop').get_text(strip=True)

    except AttributeError:
        salary = ""

    return salary


def get_skills(soup):

    try:
        div_skill = soup.find('div', class_='round_tabs_container')

        # Find all <span> elements within the <div> element
        span_elements = div_skill.find_all('span', class_='round_tabs')

        # Extract the text content from each <span> element
        skills = [span.text.strip() for span in span_elements]

    except AttributeError:
        skills = ""
    return skills

def get_jd(soup):

    try:
        div_container = soup.find('div', class_='text-container')

        # Extract the text content from the <div> element
        jd = div_container.get_text(strip=True)

    except AttributeError:
        jd = ""

    return jd

def get_app_link(soup):

    try:
        # Find the <div> element by class name
        div_with_class = new_soup.find('div', class_='buttons_container')

        # Find the <a> tag within the <div> element
        a_tag = div_with_class.find('a')

        # Extract the value of the href attribute
        app_link = "https://internshala.com" + a_tag.get('href')

    except AttributeError:
        app_link = ""

    return app_link
```

```python
if __name__ == '__main__':
  d = {"Company_name":[], "Role":[], "Location":[], "Last_date_to_apply":[],"Salary":[], "Skills":[], "Job_description":[], "Application_link":[]}
  # add your user agent
  HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})
  for i in range(1,23):
    # The webpage URL
    URL = f"https://internshala.com/jobs/business-development,data-analysis,quality-assurance-quality-control-qa-qc,software-developer-jobs/page-{i}/"

    # HTTP Request
    webpage = requests.get(URL, headers=HEADERS)

    # Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "html.parser")

    # Fetch links as List of Tag Objects
    links = soup.find_all("a", attrs={'class':'btn btn-secondary view_detail_button_outline'})

    # Store the links
    links_list = []

    # Loop for extracting links from Tag Objects
    for link in links:
      links_list.append(link.get('href'))

    for link in links_list:
          new_webpage = requests.get("https://internshala.com" + link, headers=HEADERS)

          new_soup = BeautifulSoup(new_webpage.content, "html.parser")

          # Function calls to display all necessary product information
          d['Company_name'].append(get_company_name(new_soup))
          d['Role'].append(get_role(new_soup))
          d['Location'].append(get_location(new_soup))
          d['Last_date_to_apply'].append(get_last_date(new_soup))
          d['Salary'].append(get_salary(new_soup))
          d['Skills'].append(get_skills(new_soup))
          d['Job_description'].append(get_jd(new_soup))
          d['Application_link'].append(get_app_link(new_soup))

  # Convert dictionary to DataFrame
  df = pd.DataFrame.from_dict(d)

  # Save DataFrame to CSV file
  csv_file_path = "job_data.csv"
  df.to_csv(csv_file_path, index=False)
```
